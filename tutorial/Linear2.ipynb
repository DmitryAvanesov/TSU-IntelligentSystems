```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('blood_pressure.csv')

x = df.age.values
y = df.pressure.values

plt.scatter(x, y)
```

```python
targets = (x > 25) & (x < 45)
xr = x[targets]
yr = y[targets]

plt.scatter(xr, yr)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)

model.fit(x.reshape(-1, 1), y.reshape(-1, 1))

xf = np.array([10, 80])[..., None]
yf = model.predict(xf)

plt.scatter(x, y, s=20)
# plt.scatter(xr, yr, s=25)
plt.plot(xf, yf, '-r')

# y = ax + b
a = model.coef_[0][0]
b = model.intercept_[0]

plt.title(f'y = {b:.2f} + {a:.3f}x')
```

```python
def f(x):
    return 1 * x**2 + 2 * x + 3

x = np.linspace(-2, 2)
y_ideal = np.array([f(x_) for x_ in x])

y = y_ideal + 1 * np.random.randn(y_ideal.size)

plt.scatter(x, y)

x0 = x.reshape(-1, 1)
Y = y.reshape(-1, 1)

X = np.hstack([x0, x0**2])


model = LinearRegression(fit_intercept=True)
model.fit(X, Y)

Yp = model.predict(X)
plt.plot(x, Yp, '-r')

c = model.intercept_[0]
b, a = model.coef_[0]

plt.title(f'$y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}$');
```
