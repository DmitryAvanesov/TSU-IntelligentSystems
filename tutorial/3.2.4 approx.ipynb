```python
import numpy as np
import matplotlib.pyplot as plt
```

# Самостоятельная работа по matplotlib
Придумать какую-нибудь функцию, которую будет аппроксимироваться полиномами разной степени и отрисовать на одном графике исходную зависимости и её аппроксимации полиномами разной степени.

Для аппроксимации полиномами можно использовать функцию `np.polyfit(x, y, N)`, где N - степень аппроксимирующего полинома.
Функция возвращает набор коэффициентов при x*n начиная со максимальной степени

# Пример работы:

```python
x = np.arange(-10, 10, 0.2)
y = np.sin(x)
y += np.random.normal(0, 0.2, size=y.shape)
plt.scatter(x, y, marker='x', c='k')

for N in range(1, 5):
    k = np.polyfit(x, y, N)
    f = np.poly1d(k)
    plt.plot(x, f(x), label=f'Макс. степень {N}')
    
plt.legend()
```
