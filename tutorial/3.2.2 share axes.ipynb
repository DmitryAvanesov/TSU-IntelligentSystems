```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Построение нескольких графиков на одних осях
Часто приходится строить несколько графиков на одних осях. Для того, чтобы построить двойной график, можно передать значения N функций от одного аргумента в виде матрицы с N столбцами. Рассмотрим пример где N = 2

*Примечание: здесь и далее **"_ ="** перед функциями plt нужен только чтобы подавить вывод. Можете убрать и посмотреть что получится*

```python
x = np.arange(0, 4*np.pi, 0.1)
print('x original shape:', x.shape)

# Превратим x из одномерного вектора в вектор столбец
x = x[:, np.newaxis]
print('x new shape:', x.shape)

# Тогда и эти ребята примут форму столбца
sin = np.sin(x)
cos = np.cos(x)

# А значит их можно будет стакнуть в матрицу с двумя столбцами:
stacked = np.hstack([sin, cos])
print('stacked shape:', stacked.shape)

_ = plt.plot(x, stacked)
```

Однако, такой способ не поддерживает использование стилей (см. дальше) и менее удобен, чем вызов фукнции `plt.plot()` дважды

```python
x = np.arange(0, 4*np.pi, 0.1)
plt.plot(x, np.sin(x))
_ = plt.plot(x, np.cos(x))
```

Или `plt.scatter()`

```python
x = np.arange(0, 4*np.pi, 0.1)
plt.scatter(x, np.sin(x))
_ = plt.scatter(x, np.cos(x))
```

Или даже их комбинации

```python
x = np.arange(0, 4*np.pi, 0.1)
plt.scatter(x, np.sin(x))
_ = plt.plot(x, np.cos(x))
```

## Подписи
Чтобы не запутаться в графиках можно явно задавать им стиль отрисовки и подписывать. Для подписи в функцию построения передаётся параметр `label`. После всех построений необходимо вывести легенду функцией `plt.legend()`.

```python
x = np.arange(0, 4*np.pi, 0.1)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
```

## Стили
### Тип линии

Тип линии передаётся последним параметром без аргумента в виде строки и по понятным причинам, относится только к `plt.plot()`.
matplotlib принимает следующие строки в качестве типы линии.
* **'-'** - сплошная
* **'--'** - штриховая
* **'-.'** - штрихпунктирная
* **':'** - пунктирная

Зрительные образы этих линий интуитвно понятны. Однако разместим их все на одном графике.

```python
for i, linestyle in enumerate(['-', '--', '-.', ':']):
    plt.plot(x, np.sin(x + i / 8 * np.pi), linestyle, label=linestyle)  # Обращаем внимание, что у linestyle нет названия параметра

_ = plt.legend()
```

## Цвета
Цвета передаются явно через параметр `color` и относятся ко всем функциям для построения.
Значение цвета можно передать следующим образом:
* По названию: **'blue'**, **'orange'**, **'gray'**, etc.
* Краткий код цвета (одна из букв 'rgbcmyk'). Для кого-то может стать неожиданностью, однако **'b'** - синий (ибо blue), **'k'** - чёрный (ибо key в CMYK))
* Числом **обёрнутым в строку** в интервале от **'0'** (чёрный) до **'1'** (белый). _NB: Передаётся строкой!_
* В HEX-формате: **'#FFDD44'**
* В виде RGB-кортежа. Значения кортежа - **числа** от 0 до 1: **(0.2, 0.4, 0.8)** _NB: А вот это уже передаётся числами!_
* Цвета HTML: **'orchid'** [Всех посмотреть можно, например, здесь](https://colorscheme.ru/html-colors.html)

```python
for i, color in enumerate(['orange', 'k', '0.75', '#FFDD44', (0.2, 0.4, 0.8), 'orchid']):
    plt.plot(x, np.sin(x + i / 12 * np.pi), color=color, label=str(color))
plt.legend()

_ = plt.title('Перебор методов подписи цветов')
```

# Маркеры
Для `plt.scatter()` можно передать в качестве параметра стиль маркера. Однако их гораздо больше, чем стилей маркера, поэтому их проще продемонстрировать:

```python
markers = list('oxvsd.+^<>')
for marker in markers:
    plt.scatter(np.random.rand(5), np.random.rand(5), marker=marker, label=marker)

plt.legend()
```
